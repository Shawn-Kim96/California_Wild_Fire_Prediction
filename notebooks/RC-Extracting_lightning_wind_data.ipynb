```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../data/storm_data/"
# Define file paths
details_file = data_path + "StormEvents_details-ftp_v1.0_d2024_c20250216.csv.gz"
locations_file = data_path + "StormEvents_locations-ftp_v1.0_d2024_c20250216.csv.gz"
fatalities_file = data_path + "StormEvents_fatalities-ftp_v1.0_d2024_c20250216.csv.gz"

# Load datasets
details_df = pd.read_csv(details_file, compression='gzip', low_memory=False)
locations_df = pd.read_csv(locations_file, compression='gzip', low_memory=False)
fatalities_df = pd.read_csv(fatalities_file, compression='gzip', low_memory=False)
```

```python
# Print column names
print("Details Dataset Columns:", details_df.columns.tolist())
print("Locations Dataset Columns:", locations_df.columns.tolist())
print("Fatalities Dataset Columns:", fatalities_df.columns.tolist())

# Print sample rows
print("\nDetails Sample:\n", details_df.head())
print("\nLocations Sample:\n", locations_df.head())
print("\nFatalities Sample:\n", fatalities_df.head())
```

```python
# Merge the datasets on the 'EVENT_ID' column
combined_df = details_df.merge(locations_df, on='EVENT_ID', how='left') \
                        .merge(fatalities_df, on='EVENT_ID', how='left')

# Filter for California only
california_df = combined_df[combined_df['STATE'] == 'CALIFORNIA']

# Define relevant weather types
relevant_events = ["Lightning", "High Wind", "Strong Wind", "Thunderstorm Wind", "Tornado"]

# Filter for relevant event types
filtered_CA_df = california_df[california_df["EVENT_TYPE"].isin(relevant_events)]
# Check the shape of the combined dataframe
print(f"(rows, columns) -> {filtered_CA_df.shape}")

# Save the filtered dataframe to a new CSV file
filtered_CA_df.to_csv(data_path + 'filtered_storm_data_CA_2024.csv', index=False)
```
